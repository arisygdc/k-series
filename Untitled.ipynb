```python
%store -z
from sklearn_extra.cluster import KMedoids
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
reader = pd.read_excel("data_setelah_preprocessingMeanLengkap.xlsx", skiprows=1, usecols="B:CN")
print(reader)
```

```python
# Specify starting centroid
arr = np.array(reader.iloc[:, 1:90])
mycentroid = np.array([arr[28], arr[27], arr[7]])
```

```python
kmeans = KMeans(n_clusters=3, init=mycentroid, random_state=0, n_init=1, verbose=1, max_iter=5)
kmeans.fit(arr)
print("iteration count:", kmeans.n_iter_)
centroids  = kmeans.cluster_centers_
labels = kmeans.labels_
reader['kmeans'] = labels
predictLabel = kmeans.predict(arr)
kMeandbi = davies_bouldin_score(arr, predictLabel)
print("Nilai DBI KMeans:", kMeandbi)
```

```python
print("KMeans using mean data")
print("Cluster 1")
print(reader.loc[reader['kmeans']==0, 'Kabupaten/Kota'])
print("\nCluster 2")
print(reader.loc[reader['kmeans']==1, 'Kabupaten/Kota'])
print("\nCluster 3")
print(reader.loc[reader['kmeans']==2, 'Kabupaten/Kota'])
```

```python
reader = pd.read_excel("Data_Setelah_Prepocessing_Median_K-Medoids.xlsx", skiprows=1, usecols="B:CN", sheet_name="Data Median")
# Specify starting centroid
arr = np.array(reader.iloc[:, 1:90])
mycentroid = np.array([arr[28], arr[27], arr[7]])
```

```python
kmedoids = KMedoids(n_clusters=3, init=mycentroid, random_state=0)
kmedoids.fit(arr)
labels = kmedoids.labels_
reader['kmedoids'] = labels
predictLabel = kmedoids.predict(arr)
kMedoiddbi = davies_bouldin_score(arr, predictLabel)
print("Nilai DBI KMedoids:", kMedoiddbi)
```
